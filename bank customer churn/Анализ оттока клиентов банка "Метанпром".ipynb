# Анализ оттока клиентов банка "Метанпром"

В банке «Метанпром» последнее время драматически повысился отток, это начинает становиться проблемой, которая угрожает существованию банка. Необходимо подготовить исследование об отточных клиентов банка и разработать план по их удержанию.

Информация нужна отделу маркетинга, а именно менеджеру, который занимается маркетиновыми кампаниями и работает в тесной связке с менеджерами продуктов. 

У меня есть датасет, который содержит данные о клиентах банка «Метанпром». Банк располагается в Ярославле и областных городах: Ростов Великий и Рыбинск.

Колонки:

- USERID — идентификатор пользователя (должен быть уникальным, не меняется),
- score — баллы кредитного скоринга (чем больше, тем лучше),
- city — город,
- gender — пол,
- age — возраст,
- equity — приблизительная балльная оценка собственности клиента,
- balance — баланс на счёте,
- products — количество продуктов, которыми пользуется клиент,
- credit_card — есть ли кредитная карта,
- last_activity — был ли клиент активен последнее время (за последний месяц),
- EST_SALARY — оценочная заработная плата клиента (результат работы модели)
- churn — уходит или нет (1 - клиент указал, что хочет прервать отношения с банком, 0 - нет)

**Цель исследования:** проанализировать отточных клиентов банка, чтобы потом составить рекомендации по их удержанию.

**Задачи:** 
1) подготовить датасет к анализу
2) исследовать взаимосвязи всех характеристик с оттоком разными методами, зафиксировать какие клиенты склонны уходить из банка, а какие — оставаться в нём
3) проверить статистические гипотезы
4) разработать бизнес-метрики, связанные с ценностью клиента 
5) сегментировать отточных клиентов, затем выбрать самые большие сегменты с наибольшим количеством отточных клиентов.
6) сформулировать выводы и разработать рекомендации по привлечению клиентов из приоритетных сегментов
7) подготовка презентации для отдела маркетинга

```python
from IPython.core.magic import register_line_magic
import sys
import os
from subprocess import Popen, PIPE


def is_conda_environment():
    """Return True if the current Python executable is in a conda env"""
    # TODO: make this work with Conda.exe in Windows
    conda_exec = os.path.join(os.path.dirname(sys.executable), 'conda')
    conda_history = os.path.join(sys.prefix, 'conda-meta', 'history')
    return os.path.exists(conda_exec) and os.path.exists(conda_history)


@register_line_magic
def conda(args):
    """Use conda from the current kernel"""
    # TODO: make this work with Conda.exe in Windows
    # TODO: fix string encoding to work with Python 2
    if not is_conda_environment():
        raise ValueError("The python kernel does not appear to be a conda environment.  "
                         "Please use ``%pip install`` instead.")
    
    conda_executable = os.path.join(os.path.dirname(sys.executable), 'conda')
    args = [conda_executable] + args.split()
    
    # Add --prefix to point conda installation to the current environment
    if args[1] in ['install', 'update', 'upgrade', 'remove', 'uninstall', 'list']:
        if '-p' not in args and '--prefix' not in args:
            args.insert(2, '--prefix')
            args.insert(3, sys.prefix)
            
    # Because the notebook does not allow us to respond "yes" during the
    # installation, we need to insert --yes in the argument list for some commands
    if args[1] in ['install', 'update', 'upgrade', 'remove', 'uninstall', 'create']:
        if '-y' not in args and '--yes' not in args:
            args.insert(2, '--yes')
            
    # Call conda from command line with subprocess & send results to stdout & stderr
    with Popen(args, stdout=PIPE, stderr=PIPE) as process:
        # Read stdout character by character, as it includes real-time progress updates
        for c in iter(lambda: process.stdout.read(1), b''):
            sys.stdout.write(c.decode(sys.stdout.encoding))
        # Read stderr line by line, because real-time does not matter
        for line in iter(process.stderr.readline, b''):
            sys.stderr.write(line.decode(sys.stderr.encoding))
```

Категоризируем данные из столбца с зарплатой:

```python
# формула классификации

def categorization_salary(salary):
    if (salary != salary):
        return np.nan
    #salary = int(age)
    if (salary == 0):
        return '0'
    elif (salary <= 100000):
        return '0 - 100k'
    elif (salary <= 200000):
        return '100k - 200k'
    elif (salary <= 300000):
        return '200k - 300k'
    else:
        return '300 +'

new_df['salary_category'] = new_df['estimated_salary'].apply(categorization_salary)
```

##### Заполнение пропусков в столбце balance

Мы заметили, что пропуски связаны с такими характеристиками, как заработная плата, балл кредитного скоринга, и балл оценки собственности. Но мы не можем сказать, что эти пропуски в столбце баланс на счете связаны данными характеристиками. Пропуски чаще имеются у клиентов с высоким баллом кредитного скоринга, но в то же время у них стоит нулевой балл оценки собственности. 

Вероятно, пропуски относятся к MNAR (Missing Not At Random) — механизм формирования пропусков, при котором данные отсутствуют в зависимости от неизвестных факторов. Пропуски без дополнительного обоснования нельзя отбрасывать или заполнять одним значением, т.к. это приведёт к заметным искажениям, поэтому оставим пропуски нетронутыми.

#### Загрузка  функций

```python
# функция построения сводной таблицы

def pivot_table(parameter, name):
    is_churn_table = is_churn.groupby(parameter, as_index=False).agg(count=('user_id', 'count'))
    not_churn_table = not_churn.groupby(parameter, as_index=False).agg(count=('user_id', 'count'))
    all_table = new_df.groupby(parameter, as_index=False).agg(count=('user_id', 'count'))
    
    table = all_table.merge(not_churn_table, on=parameter)
    table_data = table.merge(is_churn_table, on=parameter)
    table_data['уровень отточности %'] = round(table_data['count'] / table_data['count_x'] * 100, 2)
    table_data = table_data.rename(columns={parameter: name, 'count_x':'все клиенты', 'count_y': 'лояльные', 'count':'отточные'})
    return(table_data)
```

```python
def bar_creator(parameter, title, xaxis_title):
    table = new_df.groupby(parameter, as_index=False).agg(count=('user_id', 'count')).sort_values(by='count', 
                                                                                              ascending=False)
    table = table.reset_index(drop=True)
    fig, ax = plt.subplots(figsize=(10,5))
    ax.bar(table[parameter], table['count'])
    ax.set_ylabel('Количество клиентов')
    ax.set_xlabel(xaxis_title)
    ax.set_title(title)
    ax.bar_label(ax.containers[0], fontsize=11)

    plt.show()
```

```python
# функция построения столбикового графика с уровнями отточности по категориям

def churn_bar_creator(data, parameter, title):
    
    data = data.sort_values(by='уровень отточности %', ascending=False)
    fig, ax = plt.subplots(figsize=(10,5))
    fig = ax.bar(data[parameter], data['уровень отточности %'])
    ax.set_ylabel('Количество клиентов')
    ax.set_xlabel(parameter)
    ax.set_title(title)
    ax.set_xticks(data[parameter])
    ax.bar_label(ax.containers[0], fontsize=11)
    
    plt.axhline(y = churn_level, color ='r', linestyle='--', label='Общий уровень оттока')
    plt.legend(loc = 'upper right')
    plt.show()
```

```python
# функция построения гистаграммы

def hist_creator(parameter, bins, title, xlabel, loc='upper right'):

    is_churn = new_df[new_df['churn'] == 1][parameter]
    not_churn = new_df[new_df['churn'] == 0][parameter]
    bins = bins
    plt.hist(is_churn, bins, density=True, alpha = 0.5, label='отточные клиенты')
    plt.hist(not_churn, bins, density=True, alpha = 0.5, label='лояльные клиенты')
    plt.legend(loc=loc)
    
    plt.title(title)
    plt.xlabel(xlabel)

    plt.show()
```

## Проверка статистических гипотез

### Статистические критерии и условия их применения

**Требования к данным для примерения t-критерия Стьюдента:**

1) нормальное распределение выборочных средних
2) отстутсвие значительного числа выбросов в данных
3) равенство дисперсий

**Требования к данным для примерения u-критерия Манна-Уитни:**

1) в каждой из выборок должно быть не менее 3 значений признака. Допускается, чтобы в одной выборке было два значения, но во второй тогда не менее пяти.
2) в выборочных данных не должно быть совпадающих значений (все числа — разные) или таких совпадений должно быть очень мало (до 10).

**Условия применения статистического критерия хи-квадрата Пирсона:**
- Тип данных: параметры должны быть качественными цельночисленными частотами, измеренными в номинальной шкале, бинарными (пол: мужской/женский, наличие или отсутствие заболевания), порядковыми (степень артериальной гипертензии),
- Желательно, чтобы общее количество наблюдений было более 20,
- Ожидаемая частота, соответствующая нулевой гипотезе должна быть более 5, если ожидаемое явление принимает значение менее 5, то необходимо использовать точный Критерий Фишера.


### Проверка гипотезы "средняя зарплата клиентов, которые ушли,  различаются от средней зарплаты тех, кто остался"

Сформулируем гипотезы:

- **Нулевая гипотеза:** средняя заработная плата у отточных и у лояльных клиентов одинаковая.
- **Альтернативная гипотеза:** средняя заработная плата у отточных и у лояльных клиентов различаются.

Для проверки гипотезы используем метод Манна-Уитни, так как в данных есть большое количество выбросов.

```python
print('Средняя заработная плата у отточных клиентов:',is_churn['estimated_salary'].mean())
print('Средняя заработная плата у лояльных клиентов:', not_churn['estimated_salary'].mean())
print()
print('Медианная заработная плата у отточных клиентов:', is_churn['estimated_salary'].median())
print('Медианная заработная плата у лояльных клиентов:', not_churn['estimated_salary'].median())
```

```python
results = stats.mannwhitneyu(is_churn['estimated_salary'], not_churn['estimated_salary'], alternative='two-sided')
alpha = 0.05

print('pvalue равен', round(results.pvalue, 4))
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу ')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

По результатам проверки гипотезы u-критерием Манна-Уитни, мы можем отвергнуть нулевую гипотезу, так как, если допускаем, что верна нулевая гипотеза, вероятность того, что мы получили другой результат (что верна альтернативная гипотеза) по ошибке равен 0,003. 

Попробуем проверить те же гипотезы, но используя t-критерий Стьюдента:

```python
results = stats.ttest_ind(is_churn['estimated_salary'], not_churn['estimated_salary'], equal_var = False)
alpha = 0.05

print('pvalue равен', round(results.pvalue, 4))
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу ')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Повторим проверку гипотез t-критерием Стьюдента, но только отбросив выбросы:

```python
print('Количество клиентов в датасете:', new_df['user_id'].count())
print('Количество клиентов с зарплатой ниже 300тыс.:', new_df[new_df['estimated_salary'] <= 300000]['user_id'].count())
print('Доля выбросов:', round(
    new_df[new_df['estimated_salary'] > 300000]['user_id'].count()*100/new_df['user_id'].count(), 3))
```

```python
results = stats.ttest_ind(is_churn[is_churn['estimated_salary'] <= 300000]['estimated_salary'], 
                          not_churn[not_churn['estimated_salary'] <= 300000]['estimated_salary'], equal_var = False)
alpha = 0.05

print('pvalue равен', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу ')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

По результатам проверки гипотез t-критерием Стьюдента по исходным данным, не удалось отвергнуть нулевую гипотезу о равенстве средних. А по данным без выбросов мы получили очень низкий pvalue, что говорит о том, что если допустить, что верна нулевая гипотеза, вероятность получить другой результат, а именно, что верна альтернативная гипотеза, ничтожно мала. 

Мы убедились, что большие выбросы искажают результат при проверке гипотез, используя t-критерий Стьюдента. 

### Проверка гипотезы "средний скоринговый балл клиентов, которые ушли,  различаются от среднего балла тех, кто остался"

Сформулируем гипотезы:

- **Нулевая гипотеза:** средние скоринговие баллы у отточных и у лояльных клиентов одинаковые.
- **Альтернативная гипотеза:** средние скоринговые баллы у отточных и у лояльных клиентов различаются.

Для проверки гипотезы используем метод t-test Стьюдента, так как в данных нет больших выбросов.

```python
print('Средний балл у отточных клиентов:',is_churn['score'].mean())
print('Средний балл у лояльных клиентов:', not_churn['score'].mean())
print()
print('Медианный балл у отточных клиентов:', is_churn['score'].median())
print('Медианный балл у лояльных клиентов:', not_churn['score'].median())
```

```python
results = stats.ttest_ind(is_churn['score'], not_churn['score'], equal_var = False)
alpha = 0.05

print('pvalue равен', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу ')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Если допустить, что средние скоринговие баллы у отточных и у лояльных клиентов одинаковые, то вероятность получить другой результат, а именно то, что средние скоринговые баллы у отточных и у лояльных клиентов различаются, ниже 0.05.

- **Альтернативная гипотеза:** баланс на счету у отточных и у лояльных клиентов различаются.

Для проверки гипотезы используем метод Манна-Уитни, так как сильные есть выбросы.

```python
print('Средний баланс на счету у отточных клиентов:',is_churn['balance'].mean())
print('Средний баланс на счету у лояльных клиентов:', not_churn['balance'].mean())
print()
print('Медианный баланс на счету у отточных клиентов:', is_churn['balance'].median())
print('Медианный баланс на счету у лояльных клиентов:', not_churn['balance'].median())
```

```python
# отфильтруем данные от пропусков
is_churn_balance = is_churn[is_churn['balance'].notnull()]['balance']
not_churn_balance = not_churn[not_churn['balance'].notnull()]['balance']
```

```python
results = stats.mannwhitneyu(is_churn_balance, not_churn_balance, alternative='two-sided')
alpha = 0.05

print('pvalue равен', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу ')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Если допустить, что верна нулевая гипотеза (средний баланс на счету у отточных и у лояльных клиентов одинаковая), вероятность получить результат, что верна альтернативная гипотеза (средний баланс на счету у отточных и у лояльных клиентов различаются), очень мала, ниже 0,05.

Многие женщины в этот период жизни берут декретный отпуск. Возможно, из-за этого мужчины этого возраста больше сколнны к отточности, чем женщины. 

```python
### КОД РЕВЬЮЕРА
df = new_df
qrys = (
    'products >= 2 and balance >= 1000000',
    'last_activity == 1 and score >= 800 and salary_category == "100k - 200k" and products >=2',
    'age_category == "26 - 38" and is_man == 1 and products >= 2')

df_results = pd.concat([
    pd.DataFrame(df.query(q).agg({"churn":['mean','count']}).values,
                 columns=[q],
                 index=['Отток','Размер']).T 
    for q in qrys])

display(df_results.style.format("{:.1%}",subset=["Отток"])\
                   .format("{:_.0f}",subset=["Размер"]))
    
all_catched = pd.concat([df.query(q) for q in qrys])
all_churn = df.churn.sum()
catched = all_catched.drop_duplicates().churn.sum()
share_catched = catched / all_churn

if len(qrys) != 3:
    print('Результаты некорректны, нужно 3 лучших сегмента')
    
if df_results['Отток'].min() < df.churn.mean() * 1.5:
    print('Нужно изменить сегменты: присутствуют недостаточно отточные')
elif df_results['Отток'].min() < df.churn.mean() * 2:
    print('Часть сегментов не высокоотточна')
elif df_results['Отток'].max() > df.churn.mean() * 3:
    print('Нашли очень отточный сегмент')

if df_results['Размер'].min() < 300:
    print('Нужно изменить сегменты: присутствуют слишком мелкие сегменты')
    
if all_catched.shape[0] > 5000:
    print("Отдел маркетинга не сможет обработать такое количество клиентов")
elif all_catched.shape[0] > 3000:
    print("Отделу маркетинга будет сложно обработать такое количество клиентов, его эффективность снизится")
elif all_catched.shape[0] < 1000:
    print("Сегменты можно увеличить")

text_churn = f"{catched}/{all_churn} = {share_catched:.1%}"
if share_catched < 1/3:
    print(f"В сегменты попало недостаточное число отточных: {text_churn}")
elif share_catched < 1/2:
    print(f"В сегменты попало: {text_churn}")
elif share_catched < 2/3:
    print(f"Хороший результат, в сегменты попало: {text_churn}")
else:
    print(f"Прекрасный результат, в сегменты попало: {text_churn}")
```
